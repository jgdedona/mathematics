```python
import math
import numpy as np
```

```python
def distancecalc(x1,y1,x2,y2):
    dist = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    return dist
```

```python
print(distancecalc(1,4,3,2))
```

```python
def distance2(*points):
    x_coords = [p[0] for p in points]
    y_coords = [p[1] for p in points]
    dist = np.sqrt((x_coords[1] - x_coords[0])**2 + (y_coords[1] - y_coords[0])**2)
    return dist
```

```python
distance2((10,12), (5,6))
```

```python
def centroid2d(*points):
    x_coords = [p[0] for p in points]
    y_coords = [p[1] for p in points]
    plen = len(points)
    centroid_x = round(sum(x_coords)/plen, 2)
    centroid_y = round(sum(y_coords)/plen, 2)
    return [centroid_x, centroid_y]
```

```python
centroid2d((5,4), (1,5), (3,2), (7,9))
```

```python
def centroid3d(*points):
    x_coords = [p[0] for p in points]
    y_coords = [p[1] for p in points]
    z_coords = [p[2] for p in points]
    plen = len(points)
    centroid_x = sum(x_coords)/plen
    centroid_y = sum(y_coords)/plen
    centroid_z = sum(z_coords)/plen
    return [centroid_x, centroid_y, centroid_z]
```

```python
centroid3d((1,4,5), (4,6,5), (10,20,20))
```

```python
def k_means(*points):
    distances = []
    for point in points:
        distances.append(round(distance2(point, (2,2)), 2))
    distances2 = []
    for point in points:
        distances2.append(round(distance2(point, (8,10)), 2))
    end = [tuple(distances), tuple(distances2)]
    return end
# I want to return something along the lines of "if distance < distance2: print(cluster one) else: print (cluster two)"
```

```python
distances_tot = k_means((1,4), (3,2), (5,8), (9,10), (1,1), (0,2), (10,7))
distances_tot
```

```python
first = distances_tot[0]
second = distances_tot[1]
```

```python
for list_one,list_two in zip(first, second):
    if list_one < list_two:
        print("Cluster One")
    else:
        print("Cluster Two")
```
